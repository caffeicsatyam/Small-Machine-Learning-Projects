#EWMA :  Exponentially weighted Moving Average

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("DailyDelhiClimateTest.csv")
```

```python
df.head()
```

```python
df = df[ ['date', 'meantemp']]
df
```

```python
plt.scatter(df['date'], df['meantemp'], color='red')
```

```python
# calculating ewma

x1 = df['meantemp'].ewm(alpha=0.9).mean()
```

```python
x1
```

```python
df['ewma'] = x1
df.head()
```

```python
plt.scatter(df['date'], df['meantemp'], color='red')
plt.plot(df['date'], df['ewma'], color='blue')
plt.legend()
plt.show()
```

```python
# Variatios with changing aplha/Beta
x2 = df['meantemp'].ewm(alpha=0.5).mean()
df['ewma'] = x2

plt.scatter(df['date'], df['meantemp'], color='red')
plt.plot(df['date'], df['ewma'], color='blue')
plt.legend()
plt.show()
```

```python
# Variatios with changing aplha/Beta
x3 = df['meantemp'].ewm(alpha=0.1).mean()
df['ewma'] = x3

plt.scatter(df['date'], df['meantemp'], color='red')
plt.plot(df['date'], df['ewma'], color='blue')
plt.legend()
plt.show()
```
